# Simple linear regression

## Import the relevant libraries

```python
import numpy as np             # Importa la biblioteca NumPy para operaciones numéricas
import pandas as pd            # Importa la biblioteca Pandas para manipulación y análisis de datos
import matplotlib.pyplot as plt       # Importa la biblioteca Matplotlib para visualización de datos
import statsmodels.api as sm   # Importa la biblioteca statsmodels para análisis estadístico
import seaborn as sns          # Importa la biblioteca Seaborn para visualización de datos estadísticos
sns.set()                     # Configura los estilos predeterminados de Seaborn para trazar
```

## Load the data

```python
data = pd.read_csv('1.01. Simple linear regression.csv') # Lee el archivo CSV y carga los datos en un DataFrame de Pandas
```

```python
data.describe()  # Calcula estadísticas descriptivas para todas las columnas numéricas del DataFrame
```

# Create your first regression

## Define the dependent and the independent variables

```python
y = data['GPA']   # Se asigna la columna 'GPA' del DataFrame 'data' a la variable 'y', que representa la variable dependiente (promedio universitario de calificaciones)
x1 = data['SAT']  # Se asigna la columna 'SAT' del DataFrame 'data' a la variable 'x1', que representa la variable independiente (puntuación del SAT)
```

## Explore the data

```python
plt.scatter(x1, y)                      # Crea un gráfico de dispersión con los datos de 'x1' en el eje x y 'y' en el eje y
plt.xlabel('SAT', fontsize=20)         # Etiqueta del eje x, indica que representa la puntuación del SAT, con un tamaño de fuente de 20
plt.ylabel('GPA', fontsize=20)         # Etiqueta del eje y, indica que representa el GPA, con un tamaño de fuente de 20
plt.title('Gráfico de dispersión SAT vs GPA')  # Título del gráfico
plt.text(2400, 3.5, 'Cada punto representa un estudiante diferente', fontsize=12, color='red') # Agrega un texto explicativo en el gráfico
plt.show()                              # Muestra el gráfico
```

## Regression itself

```python
x = sm.add_constant(x1)  # Agrega una columna de unos (constantes) a la matriz de características 'x1' para el término independiente
results = sm.OLS(y, x).fit()  # Ajusta un modelo de regresión lineal ordinaria (OLS) utilizando 'y' como variable dependiente y 'x' como variable independiente
results.summary()  # Muestra un resumen detallado de los resultados del modelo de regresión lineal
```

```python
plt.scatter(x1, y)  # Crea un gráfico de dispersión con los datos de 'x1' en el eje x y 'y' en el eje y
yhat = 0.0017*x1 + 0.275  # Calcula los valores predichos de la variable dependiente (GPA) utilizando la ecuación de la regresión lineal
fig = plt.plot(x1, yhat, lw=4, c='orange', label='regression line')  # Dibuja la línea de regresión lineal en el gráfico
plt.xlabel('SAT', fontsize=20)  # Etiqueta del eje x, indica que representa la puntuación del SAT, con un tamaño de fuente de 20
plt.ylabel('GPA', fontsize=20)  # Etiqueta del eje y, indica que representa el GPA, con un tamaño de fuente de 20
plt.legend()  # Muestra la leyenda en el gráfico
plt.show()  # Muestra el gráfico
```

```python
plt.scatter(x1,y)
yhat = 0.0017*x1 + 0
fig = plt.plot(x1,yhat, lw=4, c='green', label ='regression line')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.xlim(0)
plt.ylim(0)
plt.show()
```

```python
plt.scatter(x1, y)  # Crea un gráfico de dispersión con los datos de 'x1' en el eje x y 'y' en el eje y
yhat = 0*x1 + 0.275  # Calcula los valores predichos de la variable dependiente (GPA) utilizando una predicción constante
fig = plt.plot(x1, yhat, lw=4, c='red', label='regression line')  # Dibuja la línea horizontal de predicción constante en el gráfico
plt.xlabel('SAT', fontsize=20)  # Etiqueta del eje x, indica que representa la puntuación del SAT, con un tamaño de fuente de 20
plt.ylabel('GPA', fontsize=20)  # Etiqueta del eje y, indica que representa el GPA, con un tamaño de fuente de 20
plt.legend()  # Muestra la leyenda en el gráfico
plt.show()  # Muestra el gráfico
```
